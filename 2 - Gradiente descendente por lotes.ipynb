## Problema
Ajustar una l铆nea recta $=ゐ+$ a un conjunto de datos, minimizando el error cuadr谩tico medio (MSE) entre las predicciones y los valores reales.

Este ejemplo es una forma b谩sica de inteligencia artificial porque implica que un modelo aprende autom谩ticamente una relaci贸n entre variables a partir de datos, usando una t茅cnica de optimizaci贸n para mejorar su desempe帽o.

### Importar librer铆as

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Generar datos sint茅ticos con ruido

```python
np.random.seed(42)
# Fija la semilla del generador de n煤meros aleatorios para que los resultados sean reproducibles.
# Es 煤til para garantizar que obtengas los mismos resultados cada vez que ejecutes el c贸digo.

x = np.linspace(0, 10, 150)
# Genera 150 puntos equiespaciados entre 0 y 10.
# Estos valores simulan las entradas (features) de un conjunto de datos.

y = 2.5 * x + 1.0 + np.random.normal(0, 2, size=x.shape)
# Genera los valores de salida (targets) con una relaci贸n lineal: y = 2.5x + 1.0
# A esto se le suma ruido aleatorio con distribuci贸n normal (media = 0, desviaci贸n est谩ndar = 2)
# para simular datos reales con variabilidad.
```

### Funci贸n de entrenamiento con gradiente descendente

```python
def entrenar_regresion_lineal(x, y, learning_rate, epochs):
    w = 0.0  # Inicializa el peso (pendiente de la recta) en cero
    b = 0.0  # Inicializa el sesgo/intercepto tambi茅n en cero
    n = len(x)  # N煤mero de datos de entrada
    losses = []  # Lista para registrar la p茅rdida en cada iteraci贸n

    for i in range(epochs):
        y_pred = w * x + b  # Calcula las predicciones con los par谩metros actuales
        error = y_pred - y  # Calcula el error (diferencia entre predicci贸n y valor real)

        # C谩lculo del gradiente (derivadas parciales de la funci贸n de p茅rdida con respecto a w y b)
        dw = (2/n) * np.dot(error, x)  # Derivada respecto al peso w
        db = (2/n) * np.sum(error)     # Derivada respecto al sesgo b

        # Actualiza los par谩metros con gradiente descendente
        w -= learning_rate * dw  # Nuevo valor del peso
        b -= learning_rate * db  # Nuevo valor del sesgo

        # Calcula y guarda el error cuadr谩tico medio (MSE) en esta iteraci贸n
        loss = np.mean(error ** 2)
        losses.append(loss)

        # Imprime el progreso cada 10 iteraciones o en la 煤ltima
        if i % 10 == 0 or i == epochs - 1:
            print(f"Iteraci贸n {i:3d}: w = {w:.4f}, b = {b:.4f}, Loss = {loss:.4f}")

    return w, b, losses
```

### Entrenar modelo y mostrar resultados

```python
# Entrenar modelo
w_final, b_final, historial_loss = entrenar_regresion_lineal(x, y, learning_rate=0.01, epochs=100)

# Mostrar resultados
print(f"\nModelo final: y = {w_final:.4f}x + {b_final:.4f}")
```

### Graficar resultados

```python
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.scatter(x, y, label="Datos reales")
plt.plot(x, w_final * x + b_final, color='red', label="Modelo ajustado")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Regresi贸n lineal con gradiente descendente")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(historial_loss)
plt.xlabel("Iteraci贸n")
plt.ylabel("P茅rdida (MSE)")
plt.title("Disminuci贸n del error durante el entrenamiento")
plt.grid(True)

plt.tight_layout()
plt.show()
```
